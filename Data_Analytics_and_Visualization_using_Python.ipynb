**2. Importing and Understanding Data**

**Sample Dataset**

We will use a sample dataset from Kaggle or generate synthetic data for this exercise. Let us use the famous Titanic dataset.

**Titanic Dataset:**

The RMS Titanic, a British passenger ship, sank on April 15, 1912, after colliding with an iceberg. This disaster occurred during its inaugural voyage from Southampton, England, to New York City, USA. The Titanic dataset is a well-known dataset containing information about the passengers of the ill-fated RMS Titanic. It includes various features such as:

- **PassengerId**: Unique identifier for each passenger.
- **Survived**: Whether the passenger survived (1) or not (0).
- **Pclass**: The passenger's class (1st, 2nd, or 3rd).
- **Name**: Passenger's name.
- **Sex**: Gender of the passenger.
- **Age**: Age of the passenger.
- **SibSp**: Number of siblings or spouses aboard.
- **Parch**: Number of parents or children aboard.
- **Ticket**: Ticket number.
- **Fare**: The fare the passenger paid.
- **Cabin**: Cabin number.
- **Embarked**: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton). The "Embarked" feature in the Titanic dataset indicates the port where a passenger boarded the ship. It is represented by a categorical value: **C** = Cherbourg (France), **Q** = Queenstown (Ireland) and **S** = Southampton (England). This feature helps to categorize passengers based on their embarkation points and can be useful for analyzing patterns related to passenger survival rates or socio-economic factors based on their boarding location.

The dataset is commonly used for machine learning tasks like classification.

**Code:** Importing Libraries and Data

```python
import pandas as pd  #Imports the pandas library, commonly used for data manipulation and analysis.
import numpy as np   #Imports the numpy library, which provides support for large, multi-dimensional arrays and matrices.
import matplotlib.pyplot as plt  #Imports the matplotlib library, which is used for creating static, interactive, and animated visualizations.
import seaborn as sns  #Imports the seaborn library, a powerful visualization tool built on matplotlib. It simplifies creating informative and attractive statistical plots.

# Load dataset
data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')   #The link provided points to a raw file hosted on GitHub.It is a CSV file containing the Titanic dataset.

# Display first few rows
print(data.head())

# Summary statistics
print(data.info())
#data.info() is a method used on a DataFrame to provide a concise summary of the dataset.It includes the number of entries (rows),the column names and their respective data types,the number of non-null (non-missing) entries in each column and memory usage of the DataFrame.
print(data.describe())
#data.describe() on a DataFrame provides a statistical summary of the numeric columns in the dataset.This includes: Count: The number of non-null entries. Mean, Standard deviation (std), Min (The minimum value), 25%, 50%, 75% (The 25th, 50th (median), and 75th percentiles), Max (The maximum value).
```

**3. Data Cleaning and Preprocessing**

**Steps:**

1. Handle missing values

2. Encode categorical variables

3. Feature scaling (if needed)

**Code:** Handling Missing Values

```python
# Check for missing values
print(data.isnull().sum())  #Checks for missing values in each column of the DataFrame data. It returns the count of null (missing) values per column.

# Fill missing Age with median
data['Age'].fillna(data['Age'].median(), inplace=True)  #Fills missing values in the 'Age' column of the data DataFrame with the median age value. The inplace=True modifies the original DataFrame directly.

# Drop rows with missing Embarked values
data.dropna(subset=['Embarked'], inplace=True)  #Removes rows in the data DataFrame where the 'Embarked' column has missing (null) values. The inplace=True parameter modifies the original DataFrame directly.

# Fill missing Cabin with 'Unknown'
data['Cabin'].fillna('Unknown', inplace=True)  #Replaces any missing (null) values in the 'Cabin' column of the data DataFrame with the string 'Unknown'. The inplace=True modifies the original DataFrame directly.

# Verify
print(data.isnull().sum()) #Checks and prints the number of missing (null) values in each column of the data DataFrame. It helps identify which columns have incomplete data.
```

**Code:** Encoding Categorical Variables

```python
# Encode Sex and Embarked columns
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})  #Replaces the categorical values in the 'Sex' column of the `data` DataFrame. It maps 'male' to 0 and 'female' to 1, converting the gender information into numerical form for easier analysis.
data['Embarked'] = data['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})  #Converts the categorical values in the 'Embarked' column of the data DataFrame into numerical values. 'C' is mapped to 0, 'Q' to 1, and 'S' to 2, allowing easier analysis.
```

**4. Exploratory Data Analysis (EDA)**

**Code:** Survival Rate

```python
# Survival rate
survival_rate = data['Survived'].mean() * 100   #Computes the mean of the 'Survived' column, then multiplies by 100 to express the rate as a percentage.
print(f"Survival Rate: {survival_rate:.2f}%")   #Prints the survival rate of passengers,formatted to two decimal places.It uses f-string formatting to insert the survival_rate value into the string for display,showing the percentage of survivors in the dataset.
```

Here, the survival rate is 38.25%.

**Code:** Age Distribution and Survival

```python
# Age distribution
plt.figure(figsize=(10, 6))   #Specifies the dimensions of the figure, setting the width to 10 inches and the height to 6 inches.
data['Age'].hist(bins=30, edgecolor='k') #Creates a histogram for the 'Age' column in the data DataFrame using matplotlib. It divides the data into 30 bins and adds black edges to each bin for better visual distinction.
plt.title('Age Distribution')   #Title of the plot.
plt.xlabel('Age')               #Labels the x-axis as "Age".
plt.ylabel('Frequency')         #Labels the y-axis as "Frequency".
plt.show()                      #Displays the plot.

# Age vs Survival
sns.boxplot(x='Survived', y='Age', data=data)   #Creates a boxplot showing the distribution of ages for passengers who survived (1) and did not survive (0).
plt.title('Age vs Survival')                    #Adds a title to the plot.
plt.show()                                      #Displays the plot.
```

**Age Distribution Plot Interpretation**

This histogram shows the age distribution of passengers in the dataset.

**X-axis (Age):** Represents the age of passengers, ranging from 0 to 80 years.

**Y-axis (Frequency):** Represents the number of passengers (count) in each age group.

**Bars in the Histogram:** Each bar corresponds to a specific age range (e.g., 0-2, 2-4, etc.), and the height of the bar represents how many passengers fall within that age range.

**Observations:**

1. **Peak Age Group:** The age range around 28-30 years has the highest frequency, with nearly 250 passengers in this group.

2. **Overall Shape:**
The distribution is right-skewed, meaning that most passengers were younger, with fewer passengers in older age groups.

3. **Age Clusters:**
A significant number of passengers are between the ages of 20 and 40 years.
There are also some younger passengers (children below 10) and fewer older passengers (ages 60+).

4. **Outliers:**
A small number of individuals are in the age range of 7080 years, indicating the presence of elderly passengers.

**Insights:**
1. The majority of passengers were likely working-age adults (2040 years old).
2. The presence of children (ages 010) and elderly individuals (ages 60+) suggests diversity in age groups, but they were less frequent.
3. The high concentration of passengers around 30 years of age could reflect the average age of travelers at the time.


**Box Plot Interpretation**

This box plot visualizes the relationship between Age and Survival. It shows the distribution of passenger ages for two groups: those who did not survive (0) and those who survived (1).

**X-axis (Survived):** 0: Passengers who did not survive, 1: Passengers who survived.

**Y-axis (Age):** The ages of passengers, ranging from 0 to 80+.

**Box Plot Features:**

**Median Line:** The horizontal line inside each box indicates the median (middle age).

**Interquartile Range (IQR):** The box itself spans the 25th percentile (Q1) to the 75th percentile (Q3) of ages.

**Whiskers:** The vertical lines extending from the box show the range of non-outlier ages.

**Outliers:** Small circles outside the whiskers represent extreme ages that deviate significantly from the main distribution.

**Code:** Survival by Class

```python
#Survival by Class
sns.barplot(x='Pclass', y='Survived', data=data, ci=None)    #Creates a bar plot showing the average survival rate for each passenger class (Pclass) on the Titanic. The ci=None argument disables the confidence interval shading.
plt.title('Survival Rate by Passenger Class')            #Adds a descriptive title to the plot.
plt.xlabel('Passenger Class')                                     #Passenger Class.
plt.ylabel('Survival Rate')                              #Labels the y-axis, indicating the survival rate percentage.
plt.show()                                               #Displays the plot.
```

**5. Data Visualization with Matplotlib and Seaborn**

**Code:** Correlation Heatmap

```python
# Correlation heatmap
# Ensure non-numeric columns are dropped for correlation calculation
numeric_data = data.select_dtypes(include=['number'])    #Filters the dataset to include only numeric columns.
plt.figure(figsize=(10, 6))     #Sets the plot size, setting the width to 10 inches and the height to 6 inches.
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')  #Generates a heatmap showing the correlation matrix of numeric features in the dataset. The annot=True argument adds correlation values inside the heatmap cells, cmap='coolwarm' specifies the color scheme, and fmt='.2f' formats the numbers to two decimal places.
plt.title('Correlation Heatmap')    #Sets the title.
plt.show()   #Displays the plot.
```

The output of the correlation heatmap visualization displays a matrix where each cell represents the correlation coefficient between two numeric features in the dataset.

- **Values and Coloring**:
  - Positive correlations (closer to 1) are shown in shades of red.
  - Negative correlations (closer to -1) are in shades of blue.
  - A correlation of 0 indicates no relationship, often shown as neutral colors.

- **Diagonal**: Each feature correlates perfectly with itself (value of 1), creating a diagonal of 1s.


- **Key Insights**:  
   - **Survived vs Sex (0.54)**: The correlation of **0.54** between **Survived** and **Sex** indicates a moderate positive relationship. In the Titanic dataset, 'Sex' is typically encoded as **male = 0** and **female = 1**. A positive correlation means that as the value of 'Sex' increases (from male to female), the likelihood of survival increases. This suggests that females had a higher survival rate compared to males. This aligns with historical accounts where women and children were prioritized during the evacuation of the Titanic.
   - **Pclass vs Survived (-0.34)**: In the heatmap, the correlation between **Pclass** (passenger class) and **Survived** is **-0.34**, indicating a negative correlation. This means that as the passenger class number increases (Pclass: 1 = 1st class, 2 = 2nd class, 3 = 3rd class), the survival rate decreases. Lower-class passengers (Pclass = 3) were less likely to survive compared to higher-class passengers (Pclass = 1). The negative correlation captures this inverse relationship: higher socioeconomic status (1st class) was associated with a higher chance of survival,  negative correlation, meaning lower-class passengers had lower survival rates.
   - **Fare vs Pclass (-0.55)**: Strong negative correlation as higher fares are associated with higher passenger classes.  The **-0.55 correlation** between **Fare** and **Pclass** indicates a strong negative relationship. In the Titanic dataset: **Pclass** is ordinal, where 1 = 1st class (highest), 2 = 2nd class, and 3 = 3rd class (lowest). **Fare** represents the ticket price. The negative correlation means **higher fares** are associated with **lower Pclass values** (1st class), while **lower fares** are linked to **higher Pclass values** (3rd class). In simpler terms, passengers in **1st class paid higher fares**, and those in **3rd class paid lower fares**.
   - **Parch vs SibSp (0.41)**: The positive correlation between certain variables (such as the number of family members traveling together, or features like "SibSp" and "Parch") suggests that these variables are associated with each other in a way where, as one increases, so does the other. Specifically, when examining relationships like family members traveling together, this can indicate that people with more family members aboard the ship were likely traveling together.

**Code:** Pairplot

The pairplot visualizes relationships between the features 'Survived', 'Age', 'Fare', and 'Pclass' through scatterplots for feature pairs and histograms for individual distributions. Here's what the output reveals:

1. **Diagonals**: Histograms display the distribution of each feature (e.g., age distribution, survival counts).
2. **Scatterplots**: Show interactions between features (e.g., how 'Fare' varies with 'Age' or survival).

**6. Building an Interactive Dashboard with Plotly**

**Code:** Interactive Visualization

```python
import plotly.express as px   #Imports the Plotly Express library, which is used for creating interactive plots.

# Interactive scatter plot
fig = px.scatter(data, x='Age', y='Fare', color='Survived', hover_data=['Name', 'Sex'])  #Creates an interactive scatter plot where the x-axis represents 'Age' and the y-axis represents 'Fare'. The points are color-coded based on the 'Survived' column, allowing for easy visualization of survival status. The hover_data=['Name', 'Sex'] adds additional details (Name and Sex) when hovering over points.
fig.show()      #Displays the plot.
```

The interactive scatter plot generated by the code visualizes the relationship between the 'Age' and 'Fare' columns, with points color-coded by the 'Survived' column.

Key features of the output:
1. **Axes**: The x-axis represents passengers' ages, and the y-axis shows the fare amount.
2. **Color**: Points are differentiated by survival status (e.g., distinct colors for survived or not).
3. **Hover Data**: When hovering over a point, details like the passenger's name and sex are displayed interactively.
4. **Interactivity**: Users can zoom, pan, and filter data dynamically, making it easier to analyze patterns or trends.

**Note:** The pan operation on data typically refers to the ability to move or shift the view of data within an interactive visualization without altering its contents.
